```python
import cv2
from PIL import Image
import pytesseract

def preprocess_image(image_path):
    # Load image using OpenCV
    image = cv2.imread(image_path)

    # Convert to grayscale
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    # Apply thresholding
    _, thresh = cv2.threshold(gray, 150, 255, cv2.THRESH_BINARY_INV)

    # Save the processed image (optional)
    processed_image_path = 'processed_image.png'
    cv2.imwrite(processed_image_path, thresh)

    return processed_image_path
```

```python
# Make sure to install Tesseract OCR on your system.
# Download and install from https://github.com/tesseract-ocr/tesseract

pytesseract.pytesseract.tesseract_cmd = r'C:\Program Files\Tesseract-OCR\tesseract.exe'  # Update this path as per your installation

def extract_text_from_image(image_path):
    # Preprocess the image
    processed_image_path = preprocess_image(image_path)

    # Open the image file using PIL
    img = Image.open(processed_image_path)

    # Use Tesseract to extract text
    text = pytesseract.image_to_string(img)

    return text
```

```python
import tkinter as tk
from tkinter import filedialog, Text

def open_file():
    file_path = filedialog.askopenfilename()
    if file_path:
        text = extract_text_from_image(file_path)
        text_box.delete('1.0', tk.END)
        text_box.insert(tk.END, text)

# Set up the main application window
root = tk.Tk()
root.title("Simple OCR Application")

# Create a button to open file dialog
open_button = tk.Button(root, text="Open Image", command=open_file)
open_button.pack()

# Create a text box to display the extracted text
text_box = Text(root, wrap='word')
text_box.pack(expand=1, fill='both')

# Run the application
root.mainloop()
```

```python
import cv2
from PIL import Image
import pytesseract
import tkinter as tk
from tkinter import filedialog, Text

# Set the Tesseract executable path
pytesseract.pytesseract.tesseract_cmd = r'C:\Program Files\Tesseract-OCR\tesseract.exe'

def preprocess_image(image_path):
    image = cv2.imread(image_path)
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    _, thresh = cv2.threshold(gray, 150, 255, cv2.THRESH_BINARY_INV)
    processed_image_path = 'processed_image.png'
    cv2.imwrite(processed_image_path, thresh)
    return processed_image_path

def extract_text_from_image(image_path):
    processed_image_path = preprocess_image(image_path)
    img = Image.open(processed_image_path)
    text = pytesseract.image_to_string(img)
    return text

def open_file():
    file_path = filedialog.askopenfilename()
    if file_path:
        text = extract_text_from_image(file_path)
        text_box.delete('1.0', tk.END)
        text_box.insert(tk.END, text)

root = tk.Tk()
root.title("Simple OCR Application")
open_button = tk.Button(root, text="Open Image", command=open_file)
open_button.pack()
text_box = Text(root, wrap='word')
text_box.pack(expand=1, fill='both')
root.mainloop()
```
